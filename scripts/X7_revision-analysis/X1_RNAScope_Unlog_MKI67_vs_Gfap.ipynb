```r
MW_test_mki67_combined <- function(data)
{
    grp1 <- c(data[[1]]$`C5(MKI67): Cell: Mean`, 
              data[[2]]$`C5(MKI67): Cell: Mean`,
              data[[3]]$`C5(MKI67): Cell: Mean`)
    grp2 <- c(data[[4]]$`C5(MKI67): Cell: Mean`, 
              data[[5]]$`C5(MKI67): Cell: Mean`,
              data[[6]]$`C5(MKI67): Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 7)
    
}

MW_test_mki67_individual <- function(data)
{
    grp1 <- data[[1]]$`C5(MKI67): Cell: Mean` 
    grp2 <- data[[4]]$`C5(MKI67): Cell: Mean`
    grp3 <- data[[2]]$`C5(MKI67): Cell: Mean`
    grp4 <- data[[5]]$`C5(MKI67): Cell: Mean` 
    grp5 <- data[[3]]$`C5(MKI67): Cell: Mean`
    grp6 <- data[[6]]$`C5(MKI67): Cell: Mean`
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=4.5,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 7.5))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 4.5) #+ coord_cartesian(ylim = c(0, 250))

}

MW_test_s100b_combined <- function(data)
{
    grp1 <- c(data[[1]]$`Cy3(S100b): Cell: Mean`, 
              data[[2]]$`Cy3(S100b): Cell: Mean`,
              data[[3]]$`Cy3(S100b): Cell: Mean`)
    grp2 <- c(data[[4]]$`Cy3(S100b): Cell: Mean`, 
              data[[5]]$`Cy3(S100b): Cell: Mean`,
              data[[6]]$`Cy3(S100b): Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 7)
}

MW_test_s100b_individual <- function(data)
{
    grp1 <- data[[1]]$`Cy3(S100b): Cell: Mean` 
    grp2 <- data[[4]]$`Cy3(S100b): Cell: Mean`
    grp3 <- data[[2]]$`Cy3(S100b): Cell: Mean`
    grp4 <- data[[5]]$`Cy3(S100b): Cell: Mean` 
    grp5 <- data[[3]]$`Cy3(S100b): Cell: Mean`
    grp6 <- data[[6]]$`Cy3(S100b): Cell: Mean`
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=4.5,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 7.5))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 4.5)

}

MW_test_gfap_combined <- function(data)
{
    grp1 <- c(data[[1]]$`C7(Gfap): Cell: Mean`, 
              data[[2]]$`C7(Gfap): Cell: Mean`,
              data[[3]]$`C7(Gfap): Cell: Mean`)
    grp2 <- c(data[[4]]$`C7(Gfap): Cell: Mean`, 
              data[[5]]$`C7(Gfap): Cell: Mean`,
              data[[6]]$`C7(Gfap): Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 7)
}

MW_test_gfap_individual <- function(data)
{
    grp1 <- data[[1]]$`C7(Gfap): Cell: Mean` 
    grp2 <- data[[4]]$`C7(Gfap): Cell: Mean`
    grp3 <- data[[2]]$`C7(Gfap): Cell: Mean`
    grp4 <- data[[5]]$`C7(Gfap): Cell: Mean` 
    grp5 <- data[[3]]$`C7(Gfap): Cell: Mean`
    grp6 <- data[[6]]$`C7(Gfap): Cell: Mean`
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=4.5,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 7.5))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 4.5)

}

MW_test_fitc_combined <- function(data)
{
    grp1 <- c(data[[1]]$`FITC(Cd68): Cell: Mean`, 
              data[[2]]$`FITC(Cd68): Cell: Mean`,
              data[[3]]$`FITC(Cd68): Cell: Mean`)
    grp2 <- c(data[[4]]$`FITC(Cd68): Cell: Mean`, 
              data[[5]]$`FITC(Cd68): Cell: Mean`,
              data[[6]]$`FITC(Cd68): Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size = 7)
    
}

MW_test_fitc_individual <- function(data)
{
    grp1 <- data[[1]]$`FITC(Cd68): Cell: Mean` 
    grp2 <- data[[4]]$`FITC(Cd68): Cell: Mean`
    grp3 <- data[[2]]$`FITC(Cd68): Cell: Mean`
    grp4 <- data[[5]]$`FITC(Cd68): Cell: Mean` 
    grp5 <- data[[3]]$`FITC(Cd68): Cell: Mean`
    grp6 <- data[[6]]$`FITC(Cd68): Cell: Mean`
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=4.5,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 7.5))
    P + stat_summary(fun = "mean", geom = "text", vjust = -3, aes(label = round(..y.., 2)), size=4.5)
}

MW_test_mki67_combined_ROI <- function(data)
{
    grp1 <- c(data[[1]]$`AF647_2: Cell: Mean`, 
              data[[2]]$`AF647_2: Cell: Mean`,
              data[[3]]$`AF647_2: Cell: Mean`)
    grp2 <- c(data[[4]]$`AF647_2: Cell: Mean`, 
              data[[5]]$`AF647_2: Cell: Mean`,
              data[[6]]$`AF647_2: Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = 3, aes(label = round(..y.., 2)), size = 7)
    
}

MW_test_mki67_individual_ROI <- function(data,filename)
{
    grp1 <- data[[1]]$`AF647_2: Cell: Mean` 
    grp2 <- data[[4]]$`AF647_2: Cell: Mean`
    grp3 <- data[[2]]$`AF647_2: Cell: Mean`
    grp4 <- data[[5]]$`AF647_2: Cell: Mean` 
    grp5 <- data[[3]]$`AF647_2: Cell: Mean`
    grp6 <- data[[6]]$`AF647_2: Cell: Mean`
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = 3, aes(label = round(..y.., 2)), size = 7)

}


MW_test_gfap_combined_ROI <- function(data)
{
    grp1 <- c(data[[1]]$`AF751: Cell: Mean`, 
              data[[2]]$`AF751: Cell: Mean`,
              data[[3]]$`AF751: Cell: Mean`)
    grp2 <- c(data[[4]]$`AF751: Cell: Mean`, 
              data[[5]]$`AF751: Cell: Mean`,
              data[[6]]$`AF751: Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = 3, aes(label = round(..y.., 2)), size = 7)
}

MW_test_gfap_individual_ROI <- function(data)
{
    grp1 <- c(data[[1]]$`AF751: Cell: Mean`, 
              data[[4]]$`AF751: Cell: Mean`,
              data[[2]]$`AF751: Cell: Mean`)
    grp2 <- c(data[[5]]$`AF751: Cell: Mean`, 
              data[[3]]$`AF751: Cell: Mean`,
              data[[6]]$`AF751: Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = 3, aes(label = round(..y.., 2)), size = 7)
}

MW_test_s100b_combined_ROI <- function(data)
{
    grp1 <- c(data[[1]]$`AF568: Cell: Mean`, 
              data[[2]]$`AF568: Cell: Mean`,
              data[[3]]$`AF568: Cell: Mean`)
    grp2 <- c(data[[4]]$`AF568: Cell: Mean`, 
              data[[5]]$`AF568: Cell: Mean`,
              data[[6]]$`AF568: Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2), 
          group = c(rep("Untreated", length(grp1)), rep("Treated", length(grp2))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = 3, aes(label = round(..y.., 2)), size = 7)
}

MW_test_s100b_individual_ROI <- function(data)
{
    grp1 <- c(data[[1]]$`AF568: Cell: Mean`, 
              data[[4]]$`AF568: Cell: Mean`,
              data[[2]]$`AF568: Cell: Mean`)
    grp2 <- c(data[[5]]$`AF568: Cell: Mean`, 
              data[[3]]$`AF568: Cell: Mean`,
              data[[6]]$`AF568: Cell: Mean`)
    
    df <- data.frame(value = c(grp1, grp2, grp3, grp4, grp5, grp6), 
                     group = c(rep("Bulk", length(grp1)), rep("Interface", length(grp2)), rep("Bulk", length(grp3)), 
                     rep("Interface", length(grp4)), rep("Bulk", length(grp5)), rep("Interface", length(grp6))),
                     tissue = c(rep("Tissue-1",length(grp1)+length(grp2)),rep("Tissue-2",length(grp3)+length(grp4)),
                                rep("Tissue-3",length(grp5)+length(grp6))))
    
    p <- ggboxplot(melt(df), x = "group", y = "value", color="group", palette = "jco", add = "mean", facet.by="tissue")
    P <- p + stat_compare_means(method = "wilcox.test", paired = FALSE, p.adjust.method="BH", size=7,
                           aes(label = paste0("MW - p.adj = ", after_stat(p.format)))) + theme(aspect.ratio = 1.5, 
                                                                                             text = element_text(size = 20))
    P + stat_summary(fun = "mean", geom = "text", vjust = 3, aes(label = round(..y.., 2)), size = 7)
}
```
